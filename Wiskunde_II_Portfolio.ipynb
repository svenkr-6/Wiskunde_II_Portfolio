```python
# Missende waarden per kolom met de juiste methode
mediaan_cols = ["leeftijd", "gewicht_KG", "jaarsalaris_eu", "km_auto_per_jaar",
                "km_fiets_per_jaar", "km_ov_per_jaar", "km_vliegtuig_per_jaar",
                "co2_uitstoot_per_jaar_KG", 'huishouden_grootte', "dagen_in_buitenland_per_jaar", "dagen_kantoorwerken_per_week_pro_rato", ]
modus_cols = ["geslacht", "burgerlijke_staat", "provincie",
              "opleiding_niveau", "autobezit", "fietsbezit", "elektrisch_auto",
             "gebruik_carsharing"]

for col in mediaan_cols:
    df[col] = df[col].fillna(df[col].median())
for col in modus_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

# Negatieve waarden vervangen voor mediaan
for col in ["jaarsalaris_eu", "km_auto_per_jaar", "km_ov_per_jaar",
            "km_fiets_per_jaar", "co2_uitstoot_per_jaar_KG"]:
    df.loc[df[col] < 0, col] = df[col].median()

# Onrealistische leeftijd vervangen voor mediaan
df.loc[(df["leeftijd"] < 16) | (df["leeftijd"] > 100), "leeftijd"] = df["leeftijd"].median()
```

## Hoofdstuk 2

### Transformeren

Categorische variabelen nummeren / omzetten in dummys

```python
df_transformed = df.copy()

df_transformed.replace({'geslacht': {'anders': np.nan}}, inplace=True)

mapdict = {'gebruik_carsharing':{'Nooit': 1, 'Heel soms': 2, 'Soms': 3, 'Vaak': 4, 'Altijd': 5},
           'opleiding_niveau':{'MBO': 1, 'HBO': 2, 'WO': 3},}

df_transformed = df_transformed.replace(mapdict)

binairy_cols = ['autobezit', 'fietsbezit', 'elektrisch_auto', 'geslacht', 'provincie', 'burgerlijke_staat']
df_transformed = pd.get_dummies(df_transformed, columns=binairy_cols, drop_first=True)
df_transformed = df_transformed.drop('dagen_thuiswerken_per_week_pro_rato', axis=1)
```

### Correlaties visualiseren

Correlaties tussen alle kolommen (spearman)

```python
plt.figure(figsize=(10, 12))

corr = df_transformed.corr(method='spearman')
mask= np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask)
plt.title('spearman correlatiematrix')
plt.show()
```

Nominale kolommen (Cramers V)

```python
from scipy import stats
def cramers_v(x, y):
    confusion = pd.crosstab(x, y)
    chi2 = stats.chi2_contingency(confusion)[0]
    n = confusion.values.sum()
    r, k = confusion.shape
    return np.sqrt(chi2 / (n * (min(r, k) - 1)))

dicto = {}

for i in modus_cols:
    dicto[i] = {}


    for j in modus_cols:

        dicto[i][j] = cramers_v(df[i], df[j])
        

        

dicto = pd.DataFrame(dicto)
mask = np.triu(np.ones_like(dicto))

sns.heatmap(dicto, annot=True, mask=mask, vmin=-1, vmax=1)
plt.title('cramers_v correlatiematrix')
```

Nominale kolommen met target (Eta squared)

```python
modus_cols = ['geslacht', 'burgerlijke_staat', 'provincie', 'opleiding_niveau', 'autobezit', 'fietsbezit', 'elektrisch_auto', 'gebruik_carsharing']

def eta_squared(categorisch, numeriek):
    groepen = [numeriek[categorisch == c] for c in categorisch.unique()]
    grand_mean = numeriek.mean()
    ss_between = sum(len(g) * (g.mean() - grand_mean)**2 for g in groepen)
    ss_total = ((numeriek - grand_mean)**2).sum()
    return ss_between / ss_total


corrdict = {}
for i in modus_cols:
    corrdict[i] = eta_squared(df[i], df['co2_uitstoot_per_jaar_KG'])
    
sns.heatmap(pd.DataFrame(pd.Series(corrdict).sort_values(ascending=False), 
                         columns=['co2_uitstoot_per_jaar_KG']), 
                         annot=True)

plt.title('Eta squared correlatiematrix')
plt.show()
```

## Hoofdstuk 3

### Lineair regressie model trainen

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import cross_val_score

X = df_transformed.drop(columns=["co2_uitstoot_per_jaar_KG"])
y = df_transformed["co2_uitstoot_per_jaar_KG"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,        # 20% gaat naar testset, 80% naar trainingsset
    random_state=42       # Zorgt voor reproduceerbaarheid
)
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
score = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error: {score:.2f} KG CO2 per jaar")
```

```python
# Willekeurige rij en voorspelling met formule: y = b0 + b1*x1 + b2*x2 + ...
print(f"De b0 waarde van de formule is: {model.intercept_}")
print(f"De coëfficiënten van de formule zijn: {model.coef_}")

random_rij = X_test.sample(n=1)
prediction = model.intercept_ + (model.coef_ * random_rij.values[0]).sum()
print(f"Voorspelling: {prediction:.2f} KG CO2 per jaar")
```

#### Model trainen met belangrijke features

```python
X_alle = df_transformed.drop(columns=["co2_uitstoot_per_jaar_KG"])
model_1 = LinearRegression()
#cross valideren van model_1
scores_alle = cross_val_score(model_1, X_alle, y, cv=20, scoring='neg_mean_absolute_error')
print(f"Cross-validated MAE scores: {-scores_alle}")



top5 = ["km_auto_per_jaar", "km_vliegtuig_per_jaar", "elektrisch_auto_nee",
        "dagen_in_buitenland_per_jaar", "autobezit_nee"]
X_top5 = df_transformed[top5]
model_2 = LinearRegression()
#cross valideren van model_2
scores_top5 = cross_val_score(model_2, X_top5, y, cv=20, scoring='neg_mean_absolute_error')
print(f"Cross-validated MAE scores voor top 5 features: {-scores_top5}")
```

```python
sns.lineplot(x=range(1, 21), y=-scores_alle, label='Alle features')
sns.lineplot(x=range(1, 21), y=-scores_top5, label='Top 5 features')
plt.xlabel("Cross validation fold")
plt.ylabel("MAE score")
plt.title("MAE score van alle features VS alleen top 5 features")
```

Zoals te zien is in de lijnplot hierboven is de lijn waar enkel de top 5 features in zitten hoger dan de lijn waar alle andere features ook in zitten. Dit wijst erop dat hoewel alle andere features een minder sterke correlatie met de target hebben, dat als alle features samengezet worden de MAE score lager uitkomt dan als de 5 best correlerende features op zichzelf gezet worden.

### Toetsende statistiek toepassen

```python
from scipy import stats

verschillen = scores_alle - scores_top5

# Check normaliteit van de verschillen
stat, p_norm = stats.shapiro(verschillen)

if p_norm > 0.05:
    t_stat, p_value = stats.ttest_rel(scores_alle, scores_top5)
    toets = "paired t-test"
else:
    t_stat, p_value = stats.wilcoxon(scores_alle, scores_top5)
    toets = "Wilcoxon signed-rank"

print(f"Toets: {toets}, p = {p_value:.4f}")
print(f"Gemiddelde score Model 1: {-scores_alle.mean():.4f}")
print(f"Gemiddelde score Model 2: {-scores_top5.mean():.4f}")
```

We hebben de paired t-test toegepast op de scores hierboven. We hebben de volgende denkwijze toegepast: we hebben twee groepen, dus we gebruiken een t-test. Zijn de twee groepen die we gaan toetsen onafhankelijk of gepaard? Gepaard, omdat de scores per model gekoppeld zijn aan dezelfde fold. We hebben voor twee groepen dezelfde toepassing gedaan. 

#### Toetsen interpreteren met grafieken

```python
# Histogram
plt.hist(verschillen, bins=10, edgecolor="black")
plt.title("Verdeling van de verschillen (Model 1 - Model 2)")
plt.xlabel("Verschil in score per fold")
plt.show()

# Q-Q plot: punten op de lijn = normaal verdeeld
stats.probplot(verschillen, dist="norm", plot=plt)
plt.title("Q-Q plot van de verschillen")
plt.show()
```

Door het maken van bovenstaande plotjes valt op te merken (vooral uit de histogram) dat de data niet normaal verdeeld is. 

Uit de paired t-test op de 20 CV-scores volgt p = 0.011, dus kleiner dan p = 0,05. We verwerpen hier H0; er is een significant verschil tussen beide modellen. 
Model_1 (met alle features erin) heeft de laagste MAE en presteert dus aantoonbaar beter.
